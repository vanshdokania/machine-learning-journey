```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
data = pd.read_csv(r"data\salaries-with-exp\Salary_Data.csv")

# Define features and label
X = data[["YearsExperience"]]  # Features as 1D array
y = data["Salary"]  # Target variable

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print learned parameters
print("w1 (YearsExperience):", model.coef_[0])
# print("w2 (Age):", model.coef_[1])
print("bias (intercept):", model.intercept_)

# Make predictions
y_pred = model.predict(X)

# Plotting actual vs predicted salaries (just for 1 feature visualization)
plt.scatter(X["YearsExperience"], y, color="red", label="Actual")
plt.plot(X["YearsExperience"], y_pred, color="blue", label="Predicted")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience (using sklearn LinearRegression)")
plt.legend()
plt.show()

# Load dataset
data = pd.read_csv(r"data\salaries-with-exp\Salary_Data.csv")

# Define features and label
X = data[["Age"]]  # Features as 1D array
y = data["Salary"]  # Target variable

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print learned parameters
# print("w1 (YearsExperience):", model.coef_[0])
print("w2 (Age):", model.coef_[0])
print("bias (intercept):", model.intercept_)

# Make predictions
y_pred = model.predict(X)

# Plotting actual vs predicted salaries (just for 1 feature visualization)
plt.scatter(X["Age"], y, color="red", label="Actual")
plt.plot(X["Age"], y_pred, color="blue", label="Predicted")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary vs Age (using sklearn LinearRegression)")
plt.legend()
plt.show()
```
