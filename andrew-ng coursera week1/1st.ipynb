```python
class Linear_regression():

    def __init__(self, alpha, number_of_iterations):
        
        self.alpha = alpha
        self.number_of_iterations = number_of_iterations


    def fit(self, X, Y):
        
        self.m , self.n = X.shape
        self.w = np.zeros(self.n)
        self.b = 0
        self.X = X
        self.Y = Y

        for i in range (self.number_of_iterations):
            self.update_weights()

    def update_weights(self):

        Y_predict = self.predict(self.X)
        dw = -(2*(self.X.T).dot(self.Y - Y_predict)) / self.m
        db = -2*np.sum(self.Y - Y_predict) / self.m
        self.w = self.w - self.alpha*dw
        self.b = self.b - self.alpha*db


    def predict(self, X):
        return X.dot(self.w)+self.b
    
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = salary_data.iloc[:,:-1].values
Y = salary_data.iloc[:,2].values
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.33, random_state=2)
```

```python
model = Linear_regression(alpha=0.000001,number_of_iterations=10000)
```

```python
model.fit(X_train,Y_train)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(salary_data['YearsExperience'], salary_data['Age'], salary_data['Salary'], c='red', marker='o')

# Label axes
ax.set_xlabel('Years of Experience')
ax.set_ylabel('Age')
ax.set_zlabel('Salary')

ax.set_title('3D Visualization: Salary vs Experience & Age')

plt.show()
```

```python
# Extract values
X_exp = salary_data['YearsExperience']
X_age = salary_data['Age']
Y_salary = salary_data['Salary']

# Your trained model weights
w1 = model.w[0]  # Example weight for YearsExperience
w2 = model.w[1]  # Example weight for Age
b = model.b  # Example bias (intercept)

# Create meshgrid for surface
exp_range = np.linspace(X_exp.min(), X_exp.max(), 20)
age_range = np.linspace(X_age.min(), X_age.max(), 20)
EXP, AGE = np.meshgrid(exp_range, age_range)

# Predict salary using regression equation: z = w1*x + w2*y + b
SALARY = w1 * EXP + w2 * AGE + b

# Plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter points
ax.scatter(X_exp, X_age, Y_salary, color='red', label='Data Points')

# Regression plane
ax.plot_surface(EXP, AGE, SALARY, color='blue', alpha=0.5, label='Regression Plane')

# Labels
ax.set_xlabel('Years of Experience')
ax.set_ylabel('Age')
ax.set_zlabel('Salary')
ax.set_title('Linear Regression: Salary vs Experience and Age')

plt.show()
```
