```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y,color="black", label= "data", marker='s', linewidth = 2, linestyle= "--")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = [1,2,4,5,7,8,53]
y = [1,5,0.1,24,2,12,1]
plt.bar(x, y, color = "cyan", )
plt.show()
```

```python
x = [10,13,14,19]
y = ['a','b','c','d']
plt.pie(x,label = y)
plt.show()
```
